```python
# 1. define a function "eye" returning an identity matrix with size of a given digit.

import numpy as np 

def eye(n):
    ans = [list(np.zeros(n, dtype = int)) for _ in range(n)]
    
    for i in range(n):
        ans[i][i] = 1
        
    return ans
    #return np.identity(n)

eye(5)
```

```python
#2. Linear Regression with one variable
#2.1 Plotting the Data


import pandas as pd  
import matplotlib.pyplot as plt 


data = pd.read_csv('ex1data1.txt', sep = ",", header = None)

X = data[0]
y = data[1]
m = len(y)

plt.plot(X, y, 'ro')
plt.xlabel("Population of City in 10,000s")
plt.ylabel('Profit in $10,000s')
plt.show()
```

```python
#2.2 Gradient Descent

x = np.transpose(np.array([list(np.ones(m)), list(X)]))
theta = np.array(list(np.zeros(2)))


iter = 1500
alpha = 0.01

def cost_compute(x, y, theta, m):
        
    hypo = np.matmul(x, theta)
    hypo -= y 
    cost_vec = np.matmul(np.transpose(hypo), hypo)
    cost = np.sum(cost_vec)
    
    return cost/(2*m)
    
        
    

for _ in range(iter):
    pred = np.matmul(x, theta)
    update = np.matmul(np.transpose(x), (pred - y))
    theta -= alpha * (1/m) * update
    print("Cost : ", cost_compute(x, y, theta, m))
    

    
    
```

```python
#3.1 Feature Normalisation

data2 = pd.read_csv("ex1data2.txt", sep = ',', header = None)

for i in range(len(data2.columns)):
    d_m = data2[i].mean()
    d_std = data2[i].std()
    data2[i] = (data2[i] - d_m)/d_std
    
print(data2)
    
    
```

```python
#3.2 Gradient Descent



y_2 = data2[2]
x_2 = data2.iloc[:,[0,1]]
m = len(y_2)

x_2 = pd.concat([pd.DataFrame(np.ones(m)), x_2], axis = 1)
theta_2 = np.array(list(np.zeros(3)))


iter = 1500
alpha = 0.01

def cost_compute(x, y, theta, m):
        
    hypo = np.matmul(x, theta)
    hypo -= y 
    cost_vec = np.matmul(np.transpose(hypo), hypo)
    cost = np.sum(cost_vec)
    
    return cost/(2*m)
    
        
    

for _ in range(iter):
    pred = np.matmul(x_2, theta_2)
    update = np.matmul(np.transpose(x_2), (pred - y_2))
    theta_2 -= alpha * (1/m) * update
    print("Cost : ", cost_compute(x_2, y_2, theta_2, m))
    
```
